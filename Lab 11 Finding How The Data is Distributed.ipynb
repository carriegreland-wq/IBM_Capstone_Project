### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
## Write your code here
print(f"Dataset Shape: {df.shape}")
print("-" * 30)

print("Column Information:")
df.info()

df.head()
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
## Write your code here
missing_data = df.isnull().sum()
print("Missing values per column:")
print(missing_data[missing_data > 0])
```

```python

num_cols = ['YearsCodePro', 'JobSat']
for col in num_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')
        
for col in num_cols:
    if col in df.columns:
        median_value = df[col].median()
        df[col] = df[col].fillna(median_value)
        print(f"Filled {col} with median: {median_value}")

cat_cols = ['RemoteWork', 'EdLevel', 'DevType']
for col in cat_cols:
    if col in df.columns:
        # fillna with the first value of the mode
        df[col] = df[col].fillna(df[col].mode()[0])

print("-" * 30)
print(f"Missing values remaining: {df[num_cols + cat_cols].isnull().sum().sum()}")
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
## Write your code here
key_columns = ['Employment', 'JobSat', 'YearsCodePro']

for col in key_columns:
    print(f"--- Distribution of {col} ---")
    counts = df[col].value_counts().sort_index()
    print(counts)
    print("\n")
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
## Write your code here
sns.set_theme(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='JobSat', fill=True, color='teal', bw_adjust=1)

plt.title('KDE Plot: Density of Job Satisfaction Ratings', fontsize=15)
plt.xlabel('Satisfaction Rating (1.0 - 10.0)', fontsize=12)
plt.ylabel('Density', fontsize=12)

plt.xlim(0, 10)

plt.show()
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here
def get_lang_counts(column_name):
    return df[column_name].str.split(';').explode().value_counts()

worked_counts = get_lang_counts('LanguageHaveWorkedWith').head(10)
want_counts = get_lang_counts('LanguageWantToWorkWith').head(10)

comparison_df = pd.DataFrame({
    'Worked With': worked_counts,
    'Want to Work With': want_counts
}).reset_index().rename(columns={'index': 'Language'})

comparison_melted = comparison_df.melt(id_vars='Language', var_name='Category', value_name='Count')

plt.figure(figsize=(14, 8))
sns.barplot(data=comparison_melted, x='Count', y='Language', hue='Category', palette='Set2')

plt.title('Comparison: Languages Worked With vs. Languages Wanted')
plt.xlabel('Number of Respondents')
plt.ylabel('Programming Language')
plt.legend(title='Category')
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here
top_10_countries = df['Country'].value_counts().head(10).index

df_remote = df[df['Country'].isin(top_10_countries)].copy()

remote_ct = pd.crosstab(df_remote['Country'], df_remote['RemoteWork'])

plt.figure(figsize=(12, 8))
sns.heatmap(remote_ct, annot=True, fmt="d", cmap="YlGnBu", cbar_kws={'label': 'Number of Respondents'})

plt.title('Heatmap: Remote Work Trends by Top 10 Countries')
plt.xlabel('Work Arrangement')
plt.ylabel('Country')
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
## Write your code here
pearson_corr = df['YearsCodePro'].corr(df['JobSat'], method='pearson')

print(f"Pearson Correlation Coefficient: {pearson_corr:.4f}")

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.regplot(x='YearsCodePro', y='JobSat', data=df, 
            scatter_kws={'alpha':0.1, 'color':'teal'}, 
            line_kws={'color':'red'})

plt.title('Correlation Analysis: Experience vs. Satisfaction')
plt.xlabel('Years of Professional Experience')
plt.ylabel('Job Satisfaction (1-10)')
plt.show()
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
## Write your code here
df['EdLevel_Simple'] = df['EdLevel'].str.split('(').str[0].str.strip()

top_ed = df['EdLevel_Simple'].value_counts().head(5).index
top_emp = df['Employment'].value_counts().head(5).index

ct = pd.crosstab(df['EdLevel_Simple'], df['Employment'])

ct_filtered = ct.loc[top_ed, top_emp]

ct_filtered.plot(kind='bar', stacked=True, figsize=(14, 8), colormap='viridis')

plt.title('Education Level vs. Employment Type (Top 5 Categories)')
plt.xlabel('Education Level')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Employment Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
